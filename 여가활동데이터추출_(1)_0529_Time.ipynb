```python
import seaborn as sns

# 연령대 구분 함수
def age_group(age):
    if 15 <= age <= 19:
        return '15-19'
    elif 20 <= age <= 29:
        return '20'
    elif 30 <= age <= 39:
        return '30'
    elif 40 <= age <= 49:
        return '40'
    elif 50 <= age <= 59:
        return '50'
    elif 60 <= age <= 69:
        return '60'
    else:
        return '70+'

# 연령대 칼럼 추가
df['age_group'] = df['Q34_N'].apply(age_group)

# 연령대별 데이터프레임 생성
df_15_19 = df.loc[df['age_group'] == '15-19', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_20 = df.loc[df['age_group'] == '20', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_30 = df.loc[df['age_group'] == '30', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_40 = df.loc[df['age_group'] == '40', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_50 = df.loc[df['age_group'] == '50', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_60 = df.loc[df['age_group'] == '60', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]
df_70_plus = df.loc[df['age_group'] == '70+', ['time_1', 'time_2', 'time_3', 'time_4', 'time_5']]

df_15_19
```
